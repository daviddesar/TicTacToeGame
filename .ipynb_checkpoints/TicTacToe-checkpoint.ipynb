```python
### BOARD
#Draw the board
from IPython.display import clear_output
def show_board(board):
    clear_output()
    print (board1[1] + '|' + board1[2] + '|' + board1[3])
    print (board1[4] + '|' + board1[5] + '|' + board1[6])
    print (board1[7] + '|' + board1[8] + '|' + board1[9])

### INPUT X AND O
#Player choose X or O first
def ask_XO():
    mark = ' '
    while mark != 'X' or mark != 'O':
        mark = input('Player 1 please choose your marker (X or O): ').upper()
        if mark == 'X':
            return ('X','O')
        else: return ('O', 'X')
#random for the first turn
import random
def rand():
    if random.randint(0,1) == 0:
        print ('Player 1 will go first!')
        turn = 1
    else:
        print ('Player 2 will go first!')
        turn = 2
    #return the turn=1 or turn=2


### SET UP FOR THE GAME 
#check win:
def check_win(player):
    return ((board1[1] == board1[2] == board1[3] == player) or
    (board1[4] == board1[5] == board1[6] == player) or
    (board1[7] == board1[8] == board1[9] == player) or
    (board1[1] == board1[5] == board1[9] == player) or
    (board1[1] == board1[4] == board1[7] == player) or
    (board1[2] == board1[5] == board1[8] == player) or
    (board1[3] == board1[6] == board1[9] == player) or
    (board1[3] == board1[5] == board1[7] == player))
#space check
def space_check(num):
    return board1[num] == ' '
#ask for input:
def ask_num():
    num = 0
    while num not in [1,2,3,4,5,6,7,8,9] or not space_check(num):
        num = int(input('Input number position from 1 to 9:  '))
    return num
#position = ask(num)    
def check_full():
    for i in range(1,10):
        if space_check(i) == True:
            return False
    else: return True
#place the marker:
def placer(player, position):
    board1[position] = player

def replay():
    ask_rep =''
    while ask_rep not in ['Y','N']:
        ask_rep = input('Do you want to play again? (Y/N) ').upper()
        if ask_rep == 'Y':
            return True
        else: return False

#game on
while True:
    print('Welcome to the Tic Tac Toe game! ')
    #default board 
    board1 = 10*[' ']
    show_board(board1)
    (player1,player2) = ask_XO()
    print (f"Player 1's marker: {player1}")
    print (f"Player 2's marker: {player2}")
    turn = 0
    rand()
    game_on = True
    ans = ' '
    while ans != 'Y':
        ans = input('Ready? (Y/N) ').upper()
        if ans == 'Y':
            game_on = True
    while game_on == True:
        if turn == 1:        
            print("Player 1's turn: ")
            position = ask_num()
            placer(player1, position)
            show_board(board1)
            if check_win(player1) == True:
                print ('P1 win')
                game_on = False
            elif check_full() == True:
                print ('Draw!')
                break
            else: turn = 2
        else:
            print("Player 2's turn: ")
            position = ask_num()
            placer(player2, position)
            show_board(board1)
            if check_win(player2) == True:
                print ('P2 win')
                game_on = False
            elif check_full() == True:
                print ('Draw!')
                break
            else: turn = 1
    if not replay():
        break
    
```
